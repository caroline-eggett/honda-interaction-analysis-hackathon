```python
# HOST Table

# Finding unique headlamp values (#255 invalid)
headlamp = set(df_host['headlamp'])
print("Headlamp: " + str(headlamp))

# Finding unique wiper values (#255 invalid)
wiper = set(df_host['wiper'])
print("Wiper: " + str(wiper))

# Finding unique turn signal values (#255 invalid)
turnsignal = set(df_host['turnsignal'])
print("Turn Signal: " + str(turnsignal))

# Finding unique stability control status values (#255 invalid)
stability_control_status = set(df_host['stabilitycontrolstatus'])
print("Stability Control Status: " + str(stability_control_status))

# Finding unique traction control status values (#255 invalid)
transition_state = set(df_host['transstate'])
print("Transition State: " + str(transition_state))

# Finding unique traction control status values (#255 invalid)
traction_control_status = set(df_host['traccontrolstatus'])
print("Traction Control Status: " + str(traction_control_status))

# Finding unique brake status values (#255 invalid)
brake_status = set(df_host['brakestatus'])
print("Brake Status: " + str(brake_status))

# Finding unique speed values (#1000 invalid)
speed = list(set(df_host['speed']))
for x in range(len(speed)):
    if speed[x] == 1000:
        print("Speed: " + str(speed[x]))
    else:
        pass

# Finding unique speed values (#1000 invalid)
yawrate = list(set(df_host['yawrate']))
for x in range(len(yawrate)):
    if yawrate[x] == 1000:
        print("Yaw Rate: " + str(yawrate[x]))
    else:
        pass
```
